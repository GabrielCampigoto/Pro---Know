```
import cv2
import skimage.io
from skimage.morphology import skeletonize
from skimage import util 
from google.colab import files
import matplotlib.pyplot as plt

def binary(img):
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  ret2, th2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
  return th2/255

def element():
  t = int(input('Que tipo de elemento estruturante deseja utilizar (1 para reto, 2 para elipse e 3 para cruz): '))
  k = int(input('Digite o tamanho do elemento estruturante: '))
  if t == 1:
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(k,k))
  if t == 2:
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(k,k))
  if t == 3:
    kernel = cv2.getSructuringElement(cv2.MORPH_CROSS)
  return kernel

def read_image():
  print('Entre com imagem que será trabalhada: ')
  upload = files.upload()
  path = input('Copie o caminho da imagem a qual foi realizado o upload: ')
  img = skimage.io.imread(path)
  return img

def ee_d():
  print('Necessita de um elemento estruturante, o qual é uma matriz bem definida, a qual deve ser menor que a imagem. Seu centro percorre a imagem, enquanto sua vizinhança é reponsável pela operação morfológica.')

def ans():
  res = int(input('Deseja verficar outra função: (Digite 1 para sim ou 0 para não): '))
  print('\n')
  return res

print('Este programa realiza estas funções:\n')
print('Erosão(1)\nDilatação(2)\nAbertura(3)\nFechamento(4)\nEsqueletização(5)\nGradiente erosão(6)\nGradiente dilatação(7)\nGradiente Morfológico(8)\nLimiarização por Otsu(9)\n')

res = int(input('Deseja conhecer alguma das funções acima? Digite (1) para sim e (0) para não '))

while res != 0:
    if res == 1:
        kfunction = int(input('Digite o número da função que você deseja conhecer: '))
        print('\n')
        if kfunction == 1:
            ee_d()
            print('Erosão: Ideal para separar conjuntos de pixels, diminuir conjuntos e eliminá-los, caso o elemento estruturante seja maior que o objeto na imagem.\n')
            res = ans()
            if res == 0:
                break
            else:
                res = 1
        
        if kfunction == 2:
          ee_d()
          print('Dilatação: Ideal para aumentar conjuntos, conectá-los e repara-los, caso tenha buracos entre eles.\n')
          res = ans()
          if res == 0:
            break
          else:
            res = 1
            print('\n')
          
        if kfunction == 3:
          ee_d()
          print('Abertura: É uma operação de erosão seguida de dilatação. Tem por objetivo suavizar contornos, eliminar protuberâncias e conjuntos, manter a forma do objeto original e torna a imagem mais regular, que por consequência gera uma imagem menos rica de detalhes.\n')
          res = ans()
          if res == 0:
            break
          else:
            res = 1
            print('\n')
        
        if kfunction == 4:
          ee_d()
          print('Fechamento: é uma operação de dilatação seguida de erosão. Tem por objetivo suavizar contornos, conectar conjuntos e manter os conjuntos maiores que o elemento estruturante.\n')
          res = ans()
          if res == 0:
            break
          else:
            res = 1
            print('\n')
        
        if kfunction == 5:
          print('Esqueletização: É o esqueleto de uma imagem, isso quer dizer que é uma versão afinada dos objetos presentes nela de largura de 1 pixel. Com ele é mais fácil perceber alguns padrões na imagem como nós entre cruzamentos, direções de objetos, entre outras coisas dependendo da sua área de atuação.\n')
          res = ans()
          if res == 0:
            break
          else:
            res = 1
            print('\n')
        
        if kfunction == 6:
          print('Gradiente erosão: Ideal para identificar bordas de objetos. Vem da operação Imagem original - Imagem erodida, o qual aumenta o contraste das bordas internamente ao objeto.\n')
          res = ans()
          if res == 0:
            break
          else:
            res = 1
            print('\n')

        if kfunction == 7:
          print('Gradiente dilatação: Ideal para identificar bordas de objetos. Vem da operação Imagem dilatada - Imagem original, o qual aumenta o contraste das bordas externamente ao objeto.\n')
          res = ans()
          if res == 0:
            break
          else:
            res = 1
            print('\n')

        if kfunction == 8:
          print('Gradiente morfologico: Ideal para identificar bordas. Vem da operação Imagem dilatada - Imagem erodida, destaca a transição entre os níveis de cinza de uma imagem.\n')
          res = ans()
          if res == 0:
            break
          else:
            res = 1
            print('\n')  

        if kfunction == 9:
          print('Limiarização por otsu: É um algoritmo de limiarização, o qual permite a partir de uma imagem de tons de cinza, determinar o valor ideal de limiar que permita separar os elementos do fundo e da frente em dois grupos (clusters).\n')        
          res = ans()
          if res == 0:
            break
          else:
            res = 1
            print('\n')

work = int(input('Deseja realizar alguma dessas operações (Digite 1 para sim ou 0 para não): '))
while work == 1:

  wfunction = int(input('Qual será a operação a ser realizada (digite o número da função): '))
  
  if wfunction == 1:
    img = read_image()
    kernel = element()
    erosion = cv2.erode(img,kernel,iterations = 1)
    plt.imshow(erosion,'gray')
    plt.show()
    aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
    if aa == 1:
      work = 1
    else:
      work = 0

  if wfunction == 2:
    img = read_image()
    kernel = element()
    dilatation = cv2.dilate(img,kernel,iterations = 1)
    plt.imshow(dilatation,'gray')
    plt.show()
    aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
    if aa == 1:
      work = 1
    else:
      work = 0   

  if wfunction == 3:
    img = read_image()
    kernel = element()
    opening = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
    plt.imshow(opening,'gray')
    plt.show()
    aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
    if aa == 1:
      work = 1
    else:
      work = 0

  if wfunction == 4:
    img = read_image()
    kernel = element()
    closing = cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel)
    plt.imshow(closing,'gray')
    plt.show()
    aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
    if aa == 1:
      work = 1
    else:
      work = 0

  if wfunction == 5:
    img = read_image()
    res = int(input('A sua imagem é binária (1) para sim e (0) para não: '))
    
    if res == 0:
      b = binary(img)
      plt.imshow(b,'gray')
      plt.show()
      r2 = int(input('Você deseja inverter as cores (branco para preto e preto para branco) 1 para sim e 0 para não: '))
      if r2 == 1:
        inv = 255 - img
        skeleton = skeletonize(inv)
        plt.imshow(skeleton,'gray')
        plt.show()
        aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
        if aa == 1:
          work = 1
        else:
          work = 0

      if r2 == 0:
        skeleton = skeletonize(b)
        plt.imshow(skeleton,'gray')
        plt.show()
        aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
        if aa == 1:
          work = 1
        else:
          work = 0        
    
    if res == 1:
      r2 = int(input('Você deseja inverter as cores (branco para preto e preto para branco) 1 para sim e 0 para não: '))
      if r2 == 1:
        inv = 255 - img
        skeleton = skeletonize(inv)
        plt.imshow(skeleton,'gray')
        plt.show()
        aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
        if aa == 1:
          work = 1
        else:
          work = 0

      if r2 == 0:
        skeleton = skeletonize(img)
        plt.imshow(skeleton,'gray')
        plt.show()
        aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
        if aa == 1:
          work = 1
        else:
          work = 0
  
  if wfunction == 6:
    img = read_image()
    rr = int(input('A imagem está em nível de cinza? (1) sim (0) não: '))
    if rr == 1:
      kernel = element()
      grad_ero = img - cv2.erode(img,kernel)
      plt.imshow(grad_ero,'gray')
      plt.show()
      aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
      if aa == 1:
        work = 1
      else:
        work = 0

    if rr == 0:
      gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
      kernel = element()
      grad_ero = gray - cv2.erode(gray,kernel)
      plt.imshow(grad_ero,'gray')
      plt.show()
      aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
      if aa == 1:
        work = 1
      else:
        work = 0
  
  if wfunction == 7:
    img = read_image()
    rr = int(input('A imagem está em nível de cinza? (1) sim (0) não: '))
    if rr == 1:
      kernel = element()
      grad_dil = cv2.dilate(img,kernel) - img
      plt.imshow(grad_dil,'gray')
      plt.show()
      aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
      if aa == 1:
        work = 1
      else:
        work = 0

    if rr == 0:
      gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
      kernel = element()
      grad_dil = cv2.dilate(gray,kernel) - gray
      plt.imshow(grad_dil,'gray')
      plt.show()
      aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
      if aa == 1:
        work = 1
      else:
        work = 0
  
  if wfunction == 8:
    img = read_image()
    rr = int(input('A imagem está em nível de cinza? (1) sim (0) não: '))
    if rr == 1:
      kernel = element()   
      grad_mor = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
      plt.imshow(grad_mor,'gray')
      plt.show()
      aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
      if aa == 1:
        work = 1
      else:
        work = 0

    if rr == 0:
      gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
      kernel = element()   
      grad_mor = cv2.morphologyEx(gray, cv2.MORPH_GRADIENT, kernel)
      plt.imshow(grad_mor,'gray')
      plt.show()
      aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
      if aa == 1:
        work = 1
      else:
        work = 0
  
  if wfunction == 9:
    img = read_image()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret2, th2 = cv2.threshold (gray, 0,255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    plt.imshow(th2, 'gray')
    plt.show()
    aa = int(input('Deseja realizar mais alguma operação? (1) ou (0) não: '))
    if aa == 1:
      work = 1
    else:
      work = 0

print('Agradecemos por usar o Pro-know')
 
```
